```python
#extract file from zip
import zipfile
zip_ref=zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
#read data from file and handle data from
import pandas as pd
df=pd.read_csv('orders.csv')
df.head()
```

```python
import numpy as np
df.replace(['Not Available', 'unknown'],np.nan,inplace=True)
```

```python
df.head(20)
```

```python
#rename columns name and replace it with lower case amd space with '_'
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head()
```

```python
# derived a column for caluclating the profit and sales price
df['discount']=(df['list_price']*df['discount_percent'])/100
```

```python
#df['sales_price']=df['list_price']-df['discount']
df['profit']=df['sales_price']-df['cost_price']
df.head()
```

```python
# convert order date from object data type to datetime
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
df.dtypes
```

```python
#drop unreqired columns like cost_price,list_price,discount_percentage
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df
```

```python
#load data into sql server
import sqlalchemy as sql
username = 'postgres'  # e.g., 'postgres'
password = '1234'  # e.g., 'password'
host = 'localhost'          # e.g., 'localhost'
port = '5432'               # Default PostgreSQL port
database = 'sys'  # e.g., 'testdb'
connection_string = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}'
# Create the engine
engine = sql.create_engine(connection_string)

# Test the connection
connection = engine.connect()
print("Connection established successfully")
```

```python
#load the data into sql server using append option
df.to_sql('df_orders',con=connection,index=False,if_exists='append')
```
